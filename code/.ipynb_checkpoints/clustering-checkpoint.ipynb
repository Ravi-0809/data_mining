```python
import pandas as pd
import base64
import numpy as np
import pickle
import os
import matplotlib.pyplot as plt 
import seaborn as sns

from numpy import array
```

## K-Means implementation class

```python
class K_Means:
	def __init__(self, k =3, tolerance = 0.0001, max_iterations = 300):
		self.k = k
		self.tolerance = tolerance
		self.max_iterations = max_iterations

	def fit(self, data):

		self.centroids = {}
		self.clusters = []

		#initialize the centroids, the first 'k' elements in the dataset will be our initial centroids
		for i in range(self.k):
			self.centroids[i] = data[i]

		#begin iterations
		for i in range(self.max_iterations):
			self.classes = {}
			for i in range(self.k):
				self.classes[i] = []

			#find the distance between the point and cluster; choose the nearest centroid
			for features in data:
				distances = [np.linalg.norm(features - self.centroids[centroid]) for centroid in self.centroids]
				classification = distances.index(min(distances))
				self.classes[classification].append(features)
				self.clusters.append(classification)

			previous = dict(self.centroids)

			#average the cluster datapoints to re-calculate the centroids
			for classification in self.classes:
				self.centroids[classification] = np.average(self.classes[classification], axis = 0)

			isOptimal = True

			for centroid in self.centroids:

				original_centroid = previous[centroid]
				curr = self.centroids[centroid]

				if np.sum((curr - original_centroid)/original_centroid * 100.0) > self.tolerance:
					isOptimal = False

			#break out of the main loop if the results are optimal, ie. the centroids don't change their positions much(more than our tolerance)
			if isOptimal:
				break
	def pred(self, data):
		distances = [np.linalg.norm(data - self.centroids[centroid]) for centroid in self.centroids]
		classification = distances.index(min(distances))
		return classification
```

```python
df = pd.read_csv("merged_final.csv")
```

## Selecting features for clustering

```python
df.levell.value_counts()
```

```python
df.girl_exclusive.value_counts()
```

```python
to_remove = ['Diploma', 'PG Diploma', 'Ph.D.', 'Integrated', 'M.Phil.', 'Certificate']
for degree in to_remove:
    idx = df[df['levell'] == degree].index
    df.drop(idx, inplace=True)
```

```python
df1 = df[['total_general_total', 'total_backward_castes_total']]
df2 = df[['total_general_females', 'total_backward_castes_females']]
df1.fillna(0)
df2.fillna(0)
```

```python
X1 = df1.values
X2 = df2.values
```

```python
df_modified = df
```

```python
clean_nums = {"levell" : {"Under Graduate": 0, "Post Graduate":1}}
df_modified.replace(clean_nums, inplace=True)
clean_nums = {"girl_exclusive" : {False: 0, True:1}}
df_modified.replace(clean_nums, inplace=True)
```

```python
actual_labels1 = np.array(df_modified['levell'])
actual_labels2 = np.array(df_modified['girl_exclusive'])
```

## Fitting data to the K-Means algorithm

```python
km1 = K_Means(2)
km1.fit(X1)

km2 = K_Means(2)
km2.fit(X2)
```

```python
N = X1.shape[0]
km1_labels = km1.clusters[-N:]
km1_labels = np.array(km1_labels)

km2_labels = km2.clusters[-N:]
km2_labels = np.array(km2_labels)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

## Obtaining results of clustering

### Result 1

```python
print(confusion_matrix(actual_labels1, km1_labels))
print(classification_report(actual_labels1, km1_labels))
```

### Result 2

```python
print(confusion_matrix(actual_labels2, km2_labels))
print(classification_report(actual_labels2, km2_labels))
```
